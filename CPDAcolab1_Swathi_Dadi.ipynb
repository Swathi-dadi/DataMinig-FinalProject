Below this text cell is a code cell that defines a Python function named `quicksort`. Don't worry about what the code does, we're just using it as an example here. To edit the code, just click the cell and start typing.

````
# The function `quicksort`:
#  - requires a single input (denoted by the parameter `arr` for an array)
#  - rearranges the numbers in `arr` by increasing value
#  - then returns the sorted array

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)
````

**Note**: What is the difference between `return` and `print`? And how does this affect displaying versus using the output of the code?

Google Colab makes the distinction between these two difficult to untangle  because it can automatically display the evaluation of the last line of code in a code cell. 

* `print()` is a built-in function. It takes as input any Python object and attempts to display it on the terminal. However, this is only a display; you cannot access the value of the `print()` function. 

* In comparison, `return` is a specific keyword used at the end of a function that "returns" the result as a Python object that you can then use in later lines of code via an assignment statement.

**So if you want to display your output, use `print()` or call the results on the last line of a code cell (specific to Google Colab).**

**If you need to access the output of a function, use `return` in your functions and then, after calling your function, assign the output to a variable**

### Organization of Python Code: Packages to Functions
- **Package or libraries**: a directory of 1 or more modules
  - Installed using `!pip` on the system command line
  - Example: `!pip install sklearn`
- [**Modules**](https://realpython.com/python-modules-packages/): a Python file that contains 1 or more classes or functions
  - Use `import` to bring the contents of a module into your workspace
  - Example: `import sklearn.preprocessing` or `from sklearn import preprocessing`
  - Here, we have a case of `import module.submodule`
- [**Classes**](https://realpython.com/instance-class-and-static-methods-demystified/): a set of variables and functions with a common purpose
  - Keyword ```class```
  - Example: `sklearn.preprocessing.StandardScalar()`
  - Note the presence of the parentheses
- [**Functions or Methods**](https://realpython.com/defining-your-own-python-function/): a reusable, modular piece of Python code
  - Keyword ```def```
  - Optional parameters are included within the parentheses
  - Example: ```inverse_transform(X, copy=None)```
  - Note the presence of the parentheses with the parameters `X` and `copy`
- Scripts or code snippets = single-use, simple code or code fragments


````
# This is a class. It's a Python object that acts like a container to organize 
# functions and variables with a similar purpose

class BuyCoffee:

    # It can store data in variables.
    max_shots_of_espresso = 3

    # It can contain functions, which may take input parameters (eg `num_shots`)
    def calc_cost(self, num_shots=0):
        cost = 1.25 * num_shots
        return cost

    def add_tip(amount):
        print(amount * 1.15)
  
````

### B.1 Accessing variables from imported modules

````
# This is a code snippet accessing the data from our example class `BuyCoffee`
# Note the lack of () when calling a variable `max_shots_of_espresso`

my_cafe = BuyCoffee()                # We create an instance of our class
var = my_cafe.max_shots_of_espresso  # Then we can access the variable from that class
print(var)                           # And then "do something" to that variable

# This is similar to calling the variable `pi` from the Python math module
import math         # Importing our module is similar to instantiating our class in Line 4
var = math.pi       # Then we can access the variable from that module
print(var)          # And then "do something" to that variable

````

### B.2 Accessing functions from imported modules
For now, think of functions as a black box. They may take 1 or more values as inputs, which are then transformed by the function, and then usually generates an output. 

Here, we're calling the function `calc_cost()` and supplying the input of `2` shots of espresso to get the total cost of `my_coffee`.

This is equivalent to calling the function `floor()` and supplying a decimal to get the greatest integer less than the input value as `result`.

#### Cheating cheating cheaters AKA Reading the binary (Option 3)

The preceding methods require at least two steps:
1. Download the file from the Internet to our local computer
2. Upload the file from our computer to the Google cloud

The advantage is that we have control of one copy of the data. But, if that is not a concern, the previous methods are rather cumbersome. **Why can't we import data in one step by transfering the data from the Internet directly to the Google cloud?** All data files on the Internet (and on our computers) are just a stream of binary (AKA a bunch of `1` and `0`). We just need a Python widget to get and translate the binary into its correct file format. 

Let's be very verbose so you can see what each part of the code is doing:



````
# We'll need some additional modules
import pandas as pd
import requests
import io


# Here is the URL link provided for the data set
# Note: We need to use the "raw" file link to get just the raw data
url = 'https://raw.githubusercontent.com/chaitanya2334/jupyter-tutorial/master/data/US_Accidents_May19_truncated.csv'

# Everything on the World Wide Web can be shuttled around using an "HTTP request"
response = requests.get(url)

# There are several parts to the "response", we just want the contents AKA the file
my_file = response.content

# Then, to convert the binary back to the correct file format:
my_file = io.BytesIO(my_file)

# Now this file can be read just like an uploaded file
data = pd.read_csv(my_file)
data.head()
````

````
# This can be condensed into just a few lines of code, technically 1 line if desired:
url = 'https://raw.githubusercontent.com/chaitanya2334/jupyter-tutorial/master/data/US_Accidents_May19_truncated.csv'
data = pd.read_csv(io.BytesIO(requests.get(url).content))
data.head()
````

We can use a modification of this same trick to pull data from Google Drive. 

Remember that we need a URL that points directly to the file we want, and only that file. Usually the URL will end with the file format like .csv or .doc or we can highjack a downloader link.

````
# First, we'll need a URL that points directly to the file. 

# What we have is the Google Drive share link:
my_link = 'https://drive.google.com/file/d/1OUK9Hp3vF7yu9pKiqxLpICrUUl9d9R0h/view?usp=sharing'

# Next, extract the file identifier from the link above
file_id = '1OUK9Hp3vF7yu9pKiqxLpICrUUl9d9R0h'

# Then, we're going to "borrow" the functionality of the Google downloader to 
# build a URL that delivers our file
google_dl = 'https://drive.google.com/uc?export=download&id='
url = google_dl + file_id

# Use a HTTPS request to get the file as 'response'
response = requests.get(url)

# We need to convert the raw data into a file
my_file = io.BytesIO(response.content)
pd.read_csv(my_file)
````

````
# Just like the GitHub example, the code above can be condensed into just a few lines of code:

file_id = '1OUK9Hp3vF7yu9pKiqxLpICrUUl9d9R0h'
google_dl = 'https://drive.google.com/uc?export=download&id='
data = pd.read_csv(io.BytesIO(requests.get(google_dl + file_id).content))
data.head()
````
