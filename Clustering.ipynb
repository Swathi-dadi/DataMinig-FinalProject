```python
#  Using store area, items availability and store sales value


X = df.iloc[:, [1,2,3]].values
#X = TX_df.iloc[:, [4,9]].values
#X = TX_df.iloc[:, [8,9]].values

X
```

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 30)
    kmeans.fit(X)
    # inertia method returns wcss for that model
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.lineplot(range(1, 10), wcss,marker='o',color='red')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting K-Means to the dataset - Start with 8
#kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 50)
#kmeans = KMeans(n_clusters = 3, init = 'random', random_state = 0)
y_kmeans=KMeans(n_clusters = 4, init = 'random', random_state = 30)
#y_kmeans = KMeans(n_clusters=4, init ='k-means++', max_iter=300, n_init=10,random_state=0 )
z_kmeans = y_kmeans.fit_predict(X)
```

```python
#Getting unique labels
import numpy as np
u_labels = np.unique(z_kmeans)
print (u_labels)
```

```python
plt.scatter(X[z_kmeans==0, 0], X[z_kmeans==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(X[z_kmeans==1, 0], X[z_kmeans==1, 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(X[z_kmeans==2, 0], X[z_kmeans==2, 1], s=100, c='green', label ='Cluster 3')
plt.scatter(X[z_kmeans==3, 0], X[z_kmeans==3, 1], s=100, c='cyan', label ='Cluster 4')
#plt.scatter(X[z_kmeans==4, 0], X[z_kmeans==4, 1], s=100, c='magenta', label ='Cluster 5')
#plt.scatter(X[z_kmeans==5, 0], X[z_kmeans==5, 1], s=100, c='black', label ='Cluster 6')

#Plot the centroid. This time we're going to use the cluster centres  #attribute that returns here the coordinates of the centroid.
plt.scatter(y_kmeans.cluster_centers_[:, 0], y_kmeans.cluster_centers_[:, 1], s=300, c='yellow', label = 'Centroids')
plt.title('Clusters of Sales')
plt.xlabel('Store Area  ')
plt.ylabel('Sales')
plt.show()
```
